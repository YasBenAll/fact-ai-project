```python
import pandas as pd
import os
import matplotlib.pyplot as plt

directory = "results_bounds/output"
results = os.listdir(directory)
```

```python
dfs = []

for result in results:
    df = pd.read_json(os.path.join(directory, result), orient="records")
    df["dataset"] = result[5]
    df["fairness_definition"] = result[-9:-7]
    dfs.append(df)

bounds = pd.concat(dfs)
bounds
```

```python
adult_di = bounds.loc[(bounds.dataset == "a") & (bounds.fairness_definition == "di")]
adult_dp = bounds.loc[(bounds.dataset == "a") & (bounds.fairness_definition == "dp")]
brazil_di = bounds.loc[(bounds.dataset == "b") & (bounds.fairness_definition == "di")]
brazil_dp = bounds.loc[(bounds.dataset == "b") & (bounds.fairness_definition == "dp")]
brazil_dp
```

```python
plt.plot(adult_di["interpolation_factor"], adult_di["antagonist_acc_mean"], label="Disparate Impact")
plt.fill_between(
    adult_di["interpolation_factor"],
    adult_di["antagonist_acc_mean"]-adult_di["antagonist_acc_se"], 
    adult_di["antagonist_acc_mean"]+adult_di["antagonist_acc_se"],
    alpha=0.2)
plt.plot(adult_dp["interpolation_factor"], adult_dp["antagonist_acc_mean"], label="Demographic Parity")
plt.fill_between(
    adult_dp["interpolation_factor"],
    adult_dp["antagonist_acc_mean"]-adult_dp["antagonist_acc_se"], 
    adult_dp["antagonist_acc_mean"]+adult_dp["antagonist_acc_se"],
    alpha=0.2)
plt.plot(brazil_di["interpolation_factor"], brazil_di["antagonist_acc_mean"], label="Disparate Impact")
plt.fill_between(
    brazil_di["interpolation_factor"],
    brazil_di["antagonist_acc_mean"]-brazil_di["antagonist_acc_se"], 
    brazil_di["antagonist_acc_mean"]+brazil_di["antagonist_acc_se"],
    alpha=0.2)
plt.plot(brazil_dp["interpolation_factor"], brazil_dp["antagonist_acc_mean"], label="Demographic Parity")
plt.fill_between(
    brazil_dp["interpolation_factor"],
    brazil_dp["antagonist_acc_mean"]-brazil_dp["antagonist_acc_se"], 
    brazil_dp["antagonist_acc_mean"]+brazil_dp["antagonist_acc_se"],
    alpha=0.2)
plt.title("Performance depending on bound size")
plt.xlabel("Bound size")
plt.ylabel("Classification accuracy")
plt.legend()
plt.show()
```

```python
plt.plot(adult_di["interpolation_factor"], adult_di["original_nsf"], label="Disparate Impact")
plt.fill_between(
    adult_di["interpolation_factor"],
    adult_di["original_nsf"]-adult_di["original_nsf_se"], 
    adult_di["original_nsf"]+adult_di["original_nsf_se"],
    alpha=0.2)
plt.plot(adult_dp["interpolation_factor"], adult_dp["original_nsf"], label="Demographic Parity")
plt.fill_between(
    adult_dp["interpolation_factor"],
    adult_dp["original_nsf"]-adult_dp["original_nsf_se"], 
    adult_dp["original_nsf"]+adult_dp["original_nsf_se"],
    alpha=0.2)
plt.plot(brazil_di["interpolation_factor"], brazil_di["original_nsf"], label="Disparate Impact")
plt.fill_between(
    brazil_di["interpolation_factor"],
    brazil_di["original_nsf"]-brazil_di["original_nsf_se"], 
    brazil_di["original_nsf"]+brazil_di["original_nsf_se"],
    alpha=0.2)
plt.plot(brazil_dp["interpolation_factor"], brazil_dp["original_nsf"], label="Demographic Parity")
plt.fill_between(
    brazil_dp["interpolation_factor"],
    brazil_dp["original_nsf"]-brazil_dp["original_nsf_se"], 
    brazil_dp["original_nsf"]+brazil_dp["original_nsf_se"],
    alpha=0.2)
plt.title("Proportion NSF depending on bound size")
plt.xlabel("Bound size")
plt.ylabel("Proportion NSF")
plt.legend()
plt.show()
```
